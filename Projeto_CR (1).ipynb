```python
!pip install networkx # Esta linha é um comando de terminal para instalar a biblioteca NetworkX, que é usada para criar, manipular e estudar a estrutura, dinâmica e funções de redes complexas (grafos)
#Importa as bibliotecas necessárias para o código
import networkx as nx # Para criar e manipular redes
import matplotlib.pyplot as plt # Para criar gráficos e visualizações

#------------------------------Carregando e Analisando a Rede-----------------------------------------------------------------------------------------#

# Carrega uma rede a partir de um arquivo de lista de arestas, onde cada linha representa uma aresta entre dois nós
g = nx.read_edgelist('/content/107.edges') # carregando a rede


# Imprime a ordem da rede, que é o número total de nós (ou vértices) no grafo
print('Ordem:', g.number_of_nodes())


# Imprime o número de arestas da rede, que são as conexões entre os nós
print('Número de arestas:', g.number_of_edges())


# Calcula e imprime a densidade da rede
print('Densidade:', nx.classes.function.density(g))


# Calcula e imprime o coeficiente de agrupamento médio da rede
print('Coeficiente de Agrupamento:', nx.average_clustering(g))


# Calcula e imprime o diâmetro da rede
print('Diametro', nx.diameter(g))

# Calcula e imprime o comprimento médio do caminho mais curto entre todos os pares de nós na rede
print('Grau Médio de Separação:', nx.average_shortest_path_length(g))


# Cria uma lista degrees_A contendo o grau de cada nó na rede
degrees_A = [g.degree(n) for n in g.nodes()]


# Cria um histograma dos graus dos nós na rede g para visualizar a distribuição dos graus
plt.hist(degrees_A)
# Exibe o gráfico
plt.show()
# Adiciona uma grade ao gráfico com transparência de 50%
plt.grid(alpha = 0.5)

# Armazena o coeficiente de agrupamento médio da rede em a
a= nx.average_clustering(g)

# Armazena a densidade da rede em b
b= nx.classes.function.density(g)


# Compara o coeficiente de agrupamento com a densidade.
if a > b:
  print('Coeficiente de Agrupamento é maior que a Densidade, pode-se considerar uma rede agrupada')
else:
  print('A Densidade é maior que o Coeficiente de Agrupamento, pode-se considerar uma rede não agrupada!')

#----------------------------------------Criando Rede aleatória e analisando-------------------------------------------------------------------------------#

#C ria uma árvore aleatória rt com 1034 nós
rt = nx.random_tree(1034)

# Imprime a ordem da rede, que é o número total de nós (ou vértices) no grafo
print('Ordem REDE ALEATÓRIA:', rt.number_of_nodes()) # função para ordem da rede

# Imprime o número de arestas da rede, que são as conexões entre os nós
print('Número de arestas REDE ALEATÓRIA:', rt.number_of_edges()) # função para arestes da rede

# Calcula e imprime a densidade da rede
print('Densidade REDE ALEATÓRIA:', nx.classes.function.density(rt)) # função para calcular a densidade da rede

# Calcula e imprime o coeficiente de agrupamento médio da rede
print('Coeficiente de Agrupamento REDE ALEATÓRIA:', nx.average_clustering(rt)) # função para calcular o CA da rede

# Calcula e imprime o diâmetro da rede
print('Diametro REDE ALEATÓRIA', nx.diameter(rt))

# Calcula e imprime o comprimento médio do caminho mais curto entre todos os pares de nós na rede
print('Grau Médio de Separação REDE ALEATÓRIA:', nx.average_shortest_path_length(rt))

# Cria um histograma dos graus dos nós na rede rt para visualizar a distribuição dos graus
Degrees_B = [rt.degree(n) for n in rt.nodes()]
plt.hist(Degrees_B)
plt.show()

```
